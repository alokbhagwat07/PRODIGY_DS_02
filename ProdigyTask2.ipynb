```python
# Import Important Libraries...

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Load Dataset 

dataset=pd.read_csv(r"C:\Users\ALOK BHAGWAT\Downloads\train.csv")
dataset
```

```python
# No of rows and columns 
dataset.shape
```

```python
dataset.info()
```

```python
dataset.index.name='Id No'
dataset
```

```python
dataset.describe
```

```python
# Remove Unnecessary Columns 
dataset= dataset.drop(columns=(['Cabin']))
dataset
```

```python
# Checking for null 
dataset.isnull().sum()
```

So here many null values in the age so we can fix it...

```python
# Make Columns As Meaningfull and Remove null or fix the values...
dataset['Age']=dataset['Age'].fillna(dataset['Age'].median())
dataset['Embarked']=dataset['Embarked'].fillna(dataset['Embarked'].mode()[0])
dataset
```

```python
# Finding Null Values
dataset.isnull().sum()
```

```python
# Remove Duplicates From The Dataset
dataset.drop_duplicates(inplace=True)
dataset
```

```python
# Survival Distributiom Of Data...

sns.countplot(x='Survived',data=dataset)
plt.title('Survival Distribution')
plt.show()
```

The count plot shows that around 600 passengers did not survive (0), while approximately 400 passengers survived (1). 
This indicates that a majority of passengers lost their lives, and the survival rate was lower than the mortality rate,
showing the harsh impact of the disaster.

```python
# Gender Distribution 
sns.countplot(x='Sex',data=dataset)
plt.title('The Males Vs Females')
plt.show()
```

The gender distribution plot shows that there are approximately 550 male passengers and around 300 female passengers. This indicates that the dataset is male-dominated, with males forming a significantly larger proportion of the total passengers compared to females.

```python
# Age Distribution 

sns.histplot(dataset["Age"],bins=30,kde=True)
plt.title('Age Distribution')
plt.show()
```

This Indicates That The Passenger in between 20 to 30 age is more , the age between 0 to 20 was first greater then decrease and then 
above age 40 the age is decreasing menas passenger of higher age is less...

```python
# Pclass Distribution 
sns.countplot(x='Pclass',data=dataset)
plt.title('Pclass Distribution')
plt.show()
```

The dataset is dominated by third-class passengers, with fewer first- and second-class passengers.

```python
# Survival Vs Gender Distribution

sns.countplot(x="Sex",hue='Survived',data=dataset)
plt.title('Gender vs Survival Distribution ...')
plt.show()
```

 So According to this the females are more survival than the men according to dataset...

```python
# Survival vs class distribution ...
sns.countplot(x='Pclass', hue='Survived', data=dataset)
plt.title("Survival by Passenger Class")
plt.show()
```

According to this class 1 passengers are more survival than the class 2 and 3 , in class 2 is average survival of passenger and in 
class 3 less passenger is survived...

```python
# Age vs survival distribution 
sns.boxplot(x='Survived', y='Age', data=dataset)
plt.title("Age vs Survival")
plt.show()
```

The box plot shows that the median age of survivors (1) and non-survivors (0) is quite similar, indicating that age alone was not a strong deciding factor for survival. However, survivors tend to have a slightly wider spread toward younger ages, suggesting that children and younger passengers had better chances of survival. Older age outliers are present in both groups, they have aveg survival...

```python
# Heatmap Of The Data...
plt.figure(figsize=(10,6))
sns.heatmap(dataset.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

The correlation heatmap shows that Survival is moderately influenced by Pclass and Fare. There is a negative correlation between Survived and Pclass (-0.34), indicating that passengers in lower classes (especially Class 3) had lower chances of survival.
Survival also shows a positive correlation with Fare (0.26), suggesting that passengers who paid higher fares were more likely to survive.
Age has a weak negative correlation with survival (-0.077), meaning younger passengers had a slightly higher chance of survival, but age alone was not a strong factor. 
SibSp and Parch have very weak correlations with survival, indicating family size had minimal direct impact.
Additionally, Pclass and Fare are strongly negatively correlated (-0.55), confirming that higher-class passengers generally paid higher fares.
